```python
import pickle

with open("sentences_with_entities_7548.pickle","rb") as f:
    entities = pickle.load(f)
```

```python
entity_list = []

for snt in entities:
    snt_list = []
    for entity_txt in snt.ents:
        snt_list.append(entity_txt.text)
    entity_list.append(snt_list)
```

```python
entity_list
```

```python
all_ents_similarities = []

for i in range(len(entity_list)):
    matching_ents_list = []
    for entry in entity_list:
        matching_ents = 0
        for entity in entity_list[i]:
            if entity in entry:
                matching_ents+=1
        if max(len(entry),len(entity_list[i])) > 0:
            matching_ents /= max(len(entry),len(entity_list[i]))
        matching_ents_list.append(matching_ents)
    all_ents_similarities.append(matching_ents_list)
```
